#### function for plot

```python
def plot_aot(satellite_name: str, for_date: str, table_1: pd.DataFrame, table_2: pd.DataFrame):
   
    plt.figure(dpi=300, figsize=(5.2,4.2), tight_layout=True)
    ax = plt.axes()
    plt.plot(table_1[[for_date]], label='measured', color='red', marker='.', ms=10)
    plt.plot(table_2[[for_date]], label=satellite_name, color='green', marker='.', ms=10)
    
    
    ax.grid(True, linestyle='--')
    ax.tick_params(axis='both', direction='in', width=1, length=6)
    plt.legend(title='AOT spectra')

    plt.ylabel('AOT', fontsize=14)
    plt.xlabel('Lambda, nm', fontsize=14)
    plt.title(f'AOT spectra {for_date}', fontsize=16)
    plt.savefig(f'AOT_{date}_{satellite_name}.png')
    plt.show()
```

### plot all data and save in files

```python
satellites = ['S3A', 'S3B']
dates_for_plot = ['31.05.2022', '01.08.2022', '08.08.2022']
tables = {'S3A': aot_S3A_table, 'S3B': aot_S3B_table}
```

```python
for satellite in satellites:
    for date in dates_for_plot:
        plot_aot(satellite, date, aot_measured_table, tables.get(satellite))
```
